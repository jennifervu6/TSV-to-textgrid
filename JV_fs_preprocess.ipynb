```python
import os
from pydub import AudioSegment
import numpy as np
import soundfile as sf
import librosa
```

```python
def convert_wav_mp3(input_dir, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        
    for filename in os.listdir(input_dir):
        if filename.endswith(".wav"):
            wav_file_path = os.path.join(input_dir, filename)
            mp3_filename = os.path.splitext(filename)[0] + ".mp3"
            mp3_file_path = os.path.join(output_dir, mp3_filename)

            print(f"Converting {filename}...")
            try:
                sound = AudioSegment.from_wav(wav_file_path)
                sound.export(mp3_file_path, format="mp3")
                print(f"  Saved as {mp3_filename}")
            except Exception as e:
                print(f"  Failed to convert {filename}: {e}")
```

```python
# change this part
input_folder = "wav_files"
output_folder = "mp3_files"
convert_wav_mp3(input_folder, output_folder)
```

```python
def resample(path, target_sr = 16000):

    for file in os.listdir(path):
        if file.lower().endswith(".mp3"):
            file_path = os.path.join(path, file)
            
            audio, sr = librosa.load(file_path, sr = None)
            if sr != target_sr:
                print(f'Resampling from {sr}Hz to {target_sr}Hz')
                audio_resampled = librosa.resampled(audio, original_sr = sr, target_sr = target_sr)
                sf.write(path, audio_resampled, target_sr)
            else:
                print('Audio already at target sample rate')
```

```python
path = ''
resample(path, target_sr = 16000)
```
