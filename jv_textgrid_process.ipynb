```python
import os
import pandas as pd
from praatio import textgrid
```

```python
def textgrid_convert(tsv_folder, tsv_out):
    if not os.path.exists(tsv_out):
        os.mkdir(tsv_out)

    for file in os.listdir(tsv_folder):
        if file.endswith('.tsv') or file.endswith('.csv'):
            file_path = os.path.join(tsv_folder, file)
            df = pd.read_csv(file_path, header=None, sep='\t', names=['word', 'start', 'end'])

            tg = textgrid.Textgrid()
            word_intervals = []
            for idx, row in df.iterrows():
                word_intervals.append(
                    (row['start'], row['end'], str(row['word']))
                )

            word_tier = textgrid.IntervalTier('words', word_intervals, 0, df['end'].max())
            tg.addTier(word_tier)

            # Add empty comments tier as TextTier
            comments_tier = textgrid.PointTier('comments', [], 0, df['end'].max())
            tg.addTier(comments_tier)

            out_path = os.path.join(tsv_out, file.replace('.tsv', '.TextGrid').replace('.csv', '.TextGrid'))
            tg.save(out_path, format='short_textgrid', includeBlankSpaces=True)
```

```python
input = '/Users/jennifervu/Desktop/untitled folder'
out = '/Users/jennifervu/Desktop/out'
textgrid_convert(input, out)
```
